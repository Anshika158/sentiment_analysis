```python
import matplotlib.pyplot as plt
```

```python
models = ["Logistic Regression", "SVM", "Bernoulli", "Multinomial", "NLTK"]
accuracy_scores = [82.49, 63.12, 75.98, 75.86, 85.71] 
```

```python
plt.figure(figsize=(8, 5))
plt.bar(models, accuracy_scores, color=['blue', 'green', 'red', 'purple', 'yellow'])

plt.xlabel("Models")
plt.ylabel("Accuracy Score (%)")
plt.title("Comparison of Different Models")
plt.ylim(min(accuracy_scores) - 5, max(accuracy_scores) + 5)  # Set y-axis range dynamically
plt.xticks(rotation=15)  # Rotate x-labels for better visibility
plt.show()
```
